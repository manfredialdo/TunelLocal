<h2> BACKEND JAVA </h2>

```python
# 1. Clonar repo e instalar herramientas de túnel
#!git clone https://github.com/manfredialdo/UTN-TUPaD-P3.git
!npm install -g localtunnel --quiet

import threading
import os
import time

# --- CONFI JAVA ---
# AJUSTAR ESTA RUTA al .java principal
REPO_PATH = "/content/UTN-TUPaD-P3/tp1"
MAIN_FILE = "Main.java" # NOMBRE de la clase principal
PORT = 8080             # Puerto típico de Java/Spring
# ------------------------------------

def run_java():
    os.chdir(REPO_PATH)
    print(f"compila Java en {REPO_PATH}...")
    !javac *.java
    # Ejecuta clase principal
    print(f"inicia servidor Java en puerto {PORT}...")
    !java {MAIN_FILE.replace('.java', '')}

threading.Thread(target=run_java, daemon=True).start() # 2. Ejecutar Java en segundo plano
time.sleep(5) # 3. Esperar que el servidor Java levante
# 4. Obtener IP (Password) y lanzar el tunel
print("\n" + "="*50, "\nIP PASSWORD):")
!curl -s ipv4.icanhazip.com
print("="*50,"\n","="*50,"\n", "LINK:")
!npx localtunnel --port {PORT}
```

<h2> FRONTEND: HTML CSS JS</h2>

```python
# Preparado para pruebas frontend usando maquinas virtuales...
# suponiendo no podemos instalar nada en esta pc
# 1. Preparación de ambiente, clonado y herramientas
#!git clone https://github.com/manfredialdo/UTN-TUPaD-P3.git
!pip install flask --quiet
!npm install -g localtunnel --quiet

import threading
import os
from flask import Flask, render_template_string

app = Flask(__name__) # 2. Configuración de Flask para el repo clonado
FILE_PATH = "/content/UTN-TUPaD-P3/tp1/index.html" # AjustaR la ruta al index.html dentro del repo clonado

@app.route('/')
def index():
    try:
        with open(FILE_PATH, "r", encoding="utf-8") as f:
            return render_template_string(f.read())
    except FileNotFoundError:
        return f"Error: No estA el archivo en {FILE_PATH}. VerificA la ruta.", 404

# 3. Flask en segundo plano (Puerto 5000)
def run_app():
    app.run(port=5000)

threading.Thread(target=run_app, daemon=True).start()

# 4. Obtener IP (Password) y lanzar el tunel
print("\n" + "="*50)
print("1. COPIÁ ESTA IP (ES TU TUNNEL PASSWORD):")
!curl ipv4.icanhazip.com
print("="*50)
print("2. ESPERÁ AL LINK DE ABAJO, ABRILO Y PEGÁ LA IP:")
!npx localtunnel --port 5000
```
