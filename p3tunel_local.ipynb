<h2> BACKEND PYTHON </h2>

```python
# APLICACIONES PYTHON
# 1. Preparación de ambiente, clonado y herramientas
!git clone https://github.com/manfredialdo/UTN-TUPaD-P3.git
!pip install flask --quiet
!npm install -g localtunnel --quiet

import threading
import os
from flask import Flask, render_template_string

app = Flask(__name__) # 2. Configuración de Flask para el repo clonado
FILE_PATH = "/content/UTN-TUPaD-P3/tp1/index.html" # AjustaR la ruta al index.html dentro del repo clonado

@app.route('/')
def index():
    try:
        with open(FILE_PATH, "r", encoding="utf-8") as f:
            return render_template_string(f.read())
    except FileNotFoundError:
        return f"Error: No estA el archivo en {FILE_PATH}. VerificA la ruta.", 404

# 3. Flask en segundo plano (Puerto 5000)
def run_app():
    app.run(port=5000)

threading.Thread(target=run_app, daemon=True).start()

# 4. Obtener IP (Password) y lanzar el tunel
print("\n" + "="*50)
print("1. COPIÁ ESTA IP (ES TU TUNNEL PASSWORD):")
!curl ipv4.icanhazip.com
print("="*50)
print("2. ESPERÁ AL LINK DE ABAJO, ABRILO Y PEGÁ LA IP:")
!npx localtunnel --port 5000
```
